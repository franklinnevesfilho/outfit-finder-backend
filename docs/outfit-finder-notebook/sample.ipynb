## Data Preprocessing

Now that the data is clean, I will preprocess the data:
- Encode the categorical columns
- Scale the data
- Split the data into training and testing sets
- Train the models
- Evaluate the models
- Save the models


```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])

df.head()
```

```python
feature_columns = ['gender','category', 'style', 'color', 'season']
target_column = 'usage'
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[feature_columns])
y = df[target_column]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Evaluate the model
accuracy_percentage = knn.score(X_test, y_test) * 100

accuracy_percentage
```

```python
user_input = pd.DataFrame({
    'gender':   ['Men', 'Men', 'Men', 'Men', 'Men'],
    'category': ['Topwear', 'Bottomwear','Topwear','Bottomwear', 'Bottomwear'],
    'style':    ['Tshirts', 'Jeans', 'Tshirts', 'Jeans', 'Shorts'],
    'color':    ['Black','Blue','Red', 'Black', 'Black'],
    'season':   ['Summer','Summer','Summer', 'Summer', 'Summer'],
})

for column in user_input.columns:
    user_input[column] = label_encoders[column].transform(user_input[column])
    
user_input_scaled = scaler.transform(user_input)

user_input_scaled
```

```python
# get probability of each class, and convert columns to real names

predictions = knn.predict_proba(user_input_scaled)

predictions = pd.DataFrame(predictions, columns=label_encoders['usage'].classes_)

predictions
```
